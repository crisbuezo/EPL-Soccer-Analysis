```python
epl_df['MinsPerMatch'] =(epl_df['Mins'] / epl_df['Matches']).astype(int)
epl_df['GoalsPerMatch'] = (epl_df['Goals'] / epl_df['Matches']).astype(float)
epl_df.head()
```

```python
Total_Goals = epl_df['Goals'].sum()
print(Total_Goals)
```

```python
Total_PenaltyGoals = epl_df['Penalty_Goals'].sum()
print(Total_PenaltyGoals)
```

```python
Total_PenaltyAttempts = epl_df['Penalty_Attempted'].sum()
print(Total_PenaltyAttempts)
```

```python
plt.figure(figsize=(13, 6))
pl_not_scored = epl_df['Penalty_Attempted'].sum() - Total_PenaltyGoals
data = [pl_not_scored, Total_PenaltyGoals]
labels = ['Penalties Missed', 'Penalties Scored']
color = sns.color_palette('Set2')
plt.pie(data, labels = labels, colors  = color, autopct = '%.0f%%')
plt.show()
```

```python
#players based on age groups
Under20 = epl_df[epl_df['Age'] <= 20]
age20_25 = epl_df[(epl_df['Age'] > 20) & (epl_df['Age'] <= 25)]
age25_30 = epl_df[(epl_df['Age'] > 25) & (epl_df['Age'] <= 30)]
Above30 = epl_df[epl_df['Age']>30]
```

```python
x = np.array([Under20['Name'].count(),age20_25['Name'].count(),age25_30['Name'].count(),Above30['Name'].count()])
mylabels = ["<=20", ">20 & <=25", ">25 & <=30", ">30"]
plt.title('Players by Age', fontsize = 20)
plt.pie(x, labels = mylabels, autopct="%.1f%%")
plt.show()
```

```python
#1. GRAB THE TOTAL NUMBER OF PLAYERS BY CLUB
#2. SUM OF AGES OF EACH CLUB THEN DIVIDE BY THE TOTAL NUMBER OF PLAYERS 
#3. SORT THE DATA IN DESCENDING ORDER TO GET A BETTER VISUAL ON THE DATA AND AVERAGE AGE GROUP

num_player = epl_df.groupby('Club').size()
data = (epl_df.groupby('Club')['Age'].sum())/ num_player 
data.sort_values(ascending = False)

```

```python
#Bar graph of assists by club in the entire season

Assists_by_clubs = pd.DataFrame(epl_df.groupby('Club', as_index=False)['Assists'].sum())
sns.set_theme(style="darkgrid",color_codes=False)
ax = sns.barplot(x='Club',y='Assists',data=Assists_by_clubs.sort_values(by='Assists'),palette='Set2')
ax.set_xlabel("Club",fontsize=30)
ax.set_ylabel("Assists",fontsize=20)
plt.xticks(rotation=75)
plt.rcParams['figure.figsize'] = (12,5)
plt.title('Plot of Clubs vs. Total Assists', fontsize = 25)
```

```python
#Goals by club in the entire season
Goals_by_clubs = pd.DataFrame(epl_df.groupby('Club', as_index=False)['Goals'].sum())
sns.set_theme(style="whitegrid", color_codes=True)
ax = sns.barplot(x='Club', y='Goals', data=Goals_by_clubs.sort_values(by="Goals"),palette='rocket')
ax.set_xlabel("Goals", fontsize=20)

plt.xticks(rotation=75)
plt.rcParams["figure.figsize"] = (20,8)
plt.title('Plot of Clubs vs Total Goals', fontsize=20)
```

```python
nationality_trends = epl_df.groupby('Nationality').agg(
    Total_Goals=('Goals', 'sum'),
    Total_Assists=('Assists', 'sum'),
    Players=('Name', 'count')
).sort_values(by='Total_Goals', ascending=False)
```

```python
print(nationality_trends)
```

```python
nationality_trends['Avg_Goals_Per_Player'] = nationality_trends['Total_Goals'] / nationality_trends['Players']
print(nationality_trends)
```

```python
# add columns with averages of goals and assists per nationalities.

nationality_trends['Avg_Assists_Per_Player'] = nationality_trends['Total_Assists'] / nationality_trends['Players']
```

```python
print(nationality_trends)
```

```python
nationality_trends = nationality_trends.reset_index()
```

```python
print(nationality_trends)
```

```python
top_10_nationalities = nationality_trends.head(10)
plt.figure(figsize=(10, 6))
sns.barplot(
    x='Total_Goals',
    y='Nationality',
    data=top_10_nationalities,
    palette='viridis'
)
plt.title('Top 10 Nationalities by Total Goals', fontsize=16)
plt.xlabel('Total Goals', fontsize=12)
plt.ylabel('Nationality', fontsize=12)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(
    x='Avg_Goals_Per_Player',
    y='Nationality',
    data=top_10_nationalities,
    palette='coolwarm'
    
)

plt.title('Top 10 Nationalities by Avg Goals Per Player', fontsize=16)
plt.xlabel('Ave Goals Per Player', fontsize=12)
plt.ylabel('Nationality', fontsize=12)
plt.tight_layout()
plt.show()
```
